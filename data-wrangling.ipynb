```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing data file
data = pd.read_csv(r'E-commerce Dataset.csv')
```

```python
#dropping columns that will have no influence on analysis
data = data.drop(['CustomerId', 'CustomerName', 'CustomerEmail', 'CustomerIp'], axis=1)
```

```python
data.head()
```

```python
#creating expected revenue column from ProductPrice and CostOfGood
data['ExpectedRevenue'] = data['ProductPrice'] - data['CostOfGood']
```

```python
#creating column for price paid per quantity of a good
data['PricePerQuantity'] = data['salePrice']/data['QuantitySold']
```

```python
data['DiscountPercentage'] = round((data['discountPerQuantity']/data['ProductPrice'])*100,2)
```

```python
data.head()
```

```python
data['OrderDay'] = data['OrderDay'].astype('string')
data['OrderYear'] = data['OrderYear'].astype('string')
```

```python
data['dateOfPurchase'] = data['OrderDay'] + ' ' + data['OrderMonth'] + ' ' + data['OrderYear']
```
