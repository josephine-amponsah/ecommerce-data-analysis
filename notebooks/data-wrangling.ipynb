```python
#dropping columns that will have no influence on analysis
transactions = data.drop(['CustomerName', 'CustomerEmail', 'CustomerIp'], axis=1)
```

```python
transactions.head()
```

```python
#creating expected revenue column from ProductPrice and CostOfGood
transactions['ExpectedRevenue'] = (transactions['ProductPrice'] - transactions['CostOfGood'])*transactions["QuantitySold"]
```

```python
#creating column for price paid per quantity of a good
transactions['PricePerQuantity'] = transactions['salePrice']/transactions['QuantitySold']
```

```python
transactions['discountPerQuantity'] = (transactions['discountPerQuantity'])*100
```

```python
transactions.head()
```

```python
transactions['OrderDay'] = transactions['OrderDay'].astype('string')
transactions['OrderYear'] = transactions['OrderYear'].astype('string')
```

```python
transactions['dateOfPurchase'] = transactions['OrderDay'] + ' ' + transactions['OrderMonth'] + ' ' + transactions['OrderYear']
```

```python
transactions.head()
```

```python
transactions['OrderYear'].value_counts()
```

```python
#converting time column to time data
transactions['OrderTime'] = pd.to_datetime(transactions['OrderTime'])
transactions['OrderTime'] = transactions['OrderTime'].dt.time
#data['OrderTime'] = [val.time() for val in data['OrderTime']]
```

```python
transactions.head()
```

```python
feb_data = transactions[transactions['OrderMonth']=='February']
```

```python
feb_30 = feb_data[feb_data['OrderDay'] == '30']
feb_30.head()
```

```python
#removing all rows with 3Oth February records since they vare erroneous and an insignificant number
indexes =transactions[ (transactions['OrderMonth'] == 'February') & (transactions['OrderDay'] == '30') ].index.to_list()
data_clear =transactions.drop(transactions.index[indexes], axis=0)
#indexes
```

```python
data_clear.reset_index(inplace=True)
```

```python
data_clear.info()
```

```python
non_leap = ['2017','2018','2019','2021']
index_set = data_clear[(data_clear['OrderYear'].isin(non_leap)) & (data_clear['OrderMonth'] == 'February') & (data_clear['OrderDay'] == '29')].index.to_list()
len(index_set)
```

```python
data_clear = data_clear.drop(data_clear.index[index_set], axis=0)
data_clear.info()
```

```python
index_check = data_clear[(data_clear['OrderYear'].isin(non_leap)) & (data_clear['OrderMonth'] == 'February') & (data_clear['OrderDay'] == '29')].index.to_list()
index_check
```

```python
data_clear['dateOfPurchase'] = pd.to_datetime(data_clear['dateOfPurchase'])
```

```python
data_clear = data_clear.drop("index", axis = 1)
```

```python
data_clear = data_clear.rename(columns={"ExpectedRevenue": "ExpectedProfit", "discountPerQuantity":"discountPercentage"})
```

```python
data_clear["ProfitMade"] = data_clear["salePrice"] - (data_clear["QuantitySold"]*data_clear["CostOfGood"])
```
